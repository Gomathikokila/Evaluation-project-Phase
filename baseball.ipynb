```python
x=data.drop(columns=['W','ERA'])
y=data['W']
```

Now we seperated the features and target

In Er and ERA, we removed ERA


```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=41)
```

We seperated the train and test data with the test size 30 

```python
BB=RandomForestRegressor()
BB.fit(x_train,y_train)
```

We trained the model by using RandomForestregressor method

```python
y_pred = BB.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f} ')
mae=mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae:.2f}')
```

The mean squared error value seems high, so we try to regularise it.

```python
from sklearn.linear_model import Lasso,LassoCV
```

Imported the libraries for Lasso

```python
lasscv=LassoCV(alphas=None,max_iter=10)
lasscv.fit(x_train,y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

For regularization, we got the scrore 0.76, it seems better.

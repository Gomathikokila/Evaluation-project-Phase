```python
y = data['fraud_reported']
x=data.drop(columns=['fraud_reported'])
```

seperated target and featurs

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=355)
```

Splitted train and test data

```python
def metric_score(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=clf.predict(x_train)
        print('\n==========================Train Result=====================')
        print(f'Accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%')
    elif train==False:
        pred=clf.predict(x_test)
        print('\n==========================Test Result=====================')
        print(f'Accuracy score:{accuracy_score(y_test,pred)*100:.2f}%')
        
        print('\n\n Test classifiaction report\n',classification_report(y_test,pred,digits=2))
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

trained the model with train data

```python
metric_score(clf,x_train,x_test,y_train,y_test,train=True)
metric_score(clf,x_train,x_test,y_train,y_test,train=False)
```

we achived 74 % accuracy 

By the cross validation method, we are tunning the model

```python
grid_param={
    'criterion':['gini','entropy'],
    'max_depth':range(10,15),
    'min_samples_leaf':range(2,6),
    'min_samples_split':range(3,8),
    'max_leaf_nodes':range(5,10)
}
```

```python
grid_search=GridSearchCV(estimator=clf,
                        param_grid=grid_param,
                        cv=5,
                        n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameters=grid_search.best_params_
print(best_parameters)
```

Found best params

```python
clf1=DecisionTreeClassifier(criterion= 'gini', max_depth= 10,max_leaf_nodes=5, min_samples_leaf= 2, min_samples_split= 3)

clf1.fit(x_train,y_train)
```

```python
metric_score(clf1,x_train,x_test,y_train,y_test,train=True)
metric_score(clf1,x_train,x_test,y_train,y_test,train=False)
```
